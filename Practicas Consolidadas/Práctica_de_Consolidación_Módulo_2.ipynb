```python
# Módulo 2 - Práctica de consolidación (versión menú)
# --- Sistema de Compras: Libros & Bytes ---

# 1) Inventario: lista de diccionarios (título, autor, precio, stock)
libros = [
    {"titulo": "Don Quijote de la Mancha", "autor": "Miguel de Cervantes",      "precio": 15000.0, "stock": 5},
    {"titulo": "Harry Potter y la Piedra Filosofal","autor": "J.K Rowling",     "precio": 22500.0, "stock": 2},
    {"titulo": "El Principito",             "autor": "Antoine de Saint-Exupéry","precio": 30000.0, "stock": 1},
    {"titulo": "Cien Años de Soledad",      "autor": "Gabriel Garcia Márquez",  "precio": 18900.0, "stock": 4},
    {"titulo": "El Juego de Ender",         "autor": "Orson Scott Card",        "precio": 27000.0, "stock": 3}
]

# 6) Descuentos por autor
descuentos_por_autor = {
    "J.K Rowling": 0.15,           # 15%
    "Gabriel Garcia Márquez": 0.10  # 10%
}

# Información factura
total_libros_comprados = 0
total_pagado = 0.0
total_ahorro = 0.0

# Formatos
def formatear_clp(monto):
    return f"${int(round(monto))} CLP"

# 2) Mostrar libros con stock
def mostrar_libros_disponibles():
    print("\nLibros disponibles:")
    hay = False
    for libro in libros:
        if libro["stock"] > 1:
            hay = True
            print(f"- {libro['titulo']} | Autor: {libro['autor']} | Precio: {formatear_clp(libro['precio'])} | Stock: {libro['stock']}")
    if not hay:
        print("No hay libros en stock. Intente con otro título.")

# 3) Filtrar por rango de precios
def filtrar_por_precio():
    print("\n--- Filtro por rango de precios ---")
    minimo = input("Ingrese precio mínimo: ")
    maximo = input("Ingrese precio máximo: ")
    try:
        pmin = float(minimo)
        pmax = float(maximo)
        if pmin < 0 or pmax < 0:
            print("Los precios no pueden ser negativos.")
        elif pmin > pmax:
            print("El precio mínimo no puede ser mayor que el máximo.")
        else:
            print(f"\nLibros en el rango {formatear_clp(pmin)} - {formatear_clp(pmax)}:")
            encontrados = False
            for libro in libros:
                if pmin <= libro["precio"] <= pmax:
                    encontrados = True
                    print(f"- {libro['titulo']} | Autor: {libro['autor']} | Precio: {formatear_clp(libro['precio'])} | Stock: {libro['stock']}")
            if not encontrados:
                print("No se encontraron libros en ese rango.")
    except ValueError:
        print("Valores inválidos. No se aplica filtro.")

# 4) Comprar libros: verifica existencia, stock y aplica descuento por autor
def comprar_libro():
    global total_libros_comprados, total_pagado, total_ahorro

    titulo = input("Ingrese el título del libro a comprar: ")
    cantidad_txt = input("Ingrese la cantidad deseada: ")
    try:
        cantidad = int(cantidad_txt)
    except ValueError:
        print("Cantidad inválida.")
        return

    if cantidad <= 0:
        print("La cantidad debe ser mayor que 0.")
        return

    # buscar libro por título (ignora mayúsculas/minúsculas)
    libro = None
    for item in libros:
        if item["titulo"].lower() == titulo.lower():
            libro = item
            break

    if libro is None:
        print("El libro no se encuentra en el inventario.")
        return

    if libro["stock"] < cantidad:
        print("No hay stock suficiente para completar la compra.")
        return

    subtotal = libro["precio"] * cantidad
    autor = libro["autor"]
    descuento = descuentos_por_autor.get(autor, 0.0)
    ahorro = subtotal * descuento
    total = subtotal - ahorro

    # actualizar stock y acumulados
    libro["stock"] -= cantidad
    total_libros_comprados += cantidad
    total_pagado += total
    total_ahorro += ahorro

    if descuento > 0:
        print(f"Descuento aplicado: {descuento*100:.1f}%")
    print(f"Compra exitosa: {cantidad} x {libro['titulo']} - Total: {formatear_clp(total)}")

# Menú principal con el flujo pedido
opcion = ""
while opcion != "4":
    print("\n--- Librería Libros y Bytes ---")
    print("1. Mostrar libros disponibles")
    print("2. Filtrar libros por rango de precios")
    print("3. Comprar libro")
    print("4. Finalizar compra y mostrar factura")
    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        mostrar_libros_disponibles()
    elif opcion == "2":
        filtrar_por_precio()
    elif opcion == "3":
        comprar_libro()
    elif opcion == "4":
        # 7) Factura final
        print("\n--- Factura final ---")
        print(f"Total de libros comprados: {total_libros_comprados}")
        print(f"Monto total pagado: {formatear_clp(total_pagado)}")
        print(f"Ahorro total por descuentos: {formatear_clp(total_ahorro)}")
        print("Gracias por su compra.")
    else:
        print("Opción inválida.")
```
